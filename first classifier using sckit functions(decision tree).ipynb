```python
from sklearn import datasets
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
iris= datasets.load_iris()
x=iris.data
y=iris.target
```

* importing iris dataset for classifying 
* x= input data for iris
* y= output labels
* so overall function would be like f(x)=y

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

* splitting data in train and test

```python
from sklearn import tree
classify=tree.DecisionTreeClassifier()
```

```python
classify.fit(x_train, y_train)
prediction=classify.predict(x_test)
print prediction

dot_data = StringIO()

export_graphviz(classify, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

* Above cell is for classifier

```python
from sklearn.metrics import accuracy_score
print accuracy_score(y_test, prediction)
```

* printing accuracy
